```python
# Logic function to determine the amount of commissions that will be paid for each deal, every commission will be paid at a tier one rate. if above tier 2 attainment then they will recive 
# a tier 2 bonus which is 50% of base rate and if they cross tier 3 it will be another 50% of base rate. if we take half of the commissionable ARR and multiply it by the base rate it would be the same
# as finding 50% of the base rate
def getCommissionAmount(Accumulative_ARR, Commissionable_ARR, Attainment, Rate):
    upper_band_goal = Attainment * 1.5
    tier2_diff = 0.5
    tier3_diff = 0.5

    tier1_payout = 1 * Commissionable_ARR
    tier2_payout = max(0, min(Accumulative_ARR - Attainment, Commissionable_ARR)) * tier2_diff
    tier3_payout = max(0, min(Accumulative_ARR - upper_band_goal, Commissionable_ARR)) * tier3_diff

    return Rate * (tier1_payout + tier2_payout + tier3_payout)
```

```python
#Creating the empty row for commission amount
captivateiq_Commission_Detail_df["commission_earned"] = pd.NaT
```

```python
#Loop through the data to determine the commission amount
for index, row in captivateiq_Commission_Detail_df.iterrows():
    captivateiq_Commission_Detail_df.loc[index, "commission_earned"] = getCommissionAmount(row["Accumulative ARR"], row["Commissionable ARR"], row["Attainment"], row["Rate"])
captivateiq_Commission_Detail_df.head()
```

```python
# Creating summary attainment by quarter 
captivateiq_Attainment_by_quarter_Summary_df = captivateiq_Commission_Detail_df.groupby(["Name", "attainment_quarter",])["Commissionable ARR"].sum().to_frame("attainment_by_quarter").reset_index()
captivateiq_Attainment_by_quarter_Summary_df.head(10)
```

```python
#creating payment amount by quarter
captivateiq_paid_Commission_Summary_df = captivateiq_Commission_Detail_df.groupby(["Name", "payment_quarter",])["commission_earned"].sum().to_frame("paid_by_quarter").reset_index()
captivateiq_paid_Commission_Summary_df.head(10)
```

```python
#Creating earned commission by quarter
captivateiq_earned_Commission_Summary_df = captivateiq_Commission_Detail_df.groupby(["Name", "attainment_quarter",])["commission_earned"].sum().to_frame("commission_earned_by_quarter").reset_index()
captivateiq_earned_Commission_Summary_df.head(10)
```

```python
#combine Dataframe enroute to final summary dataframe
captivateiq_Attainment_by_quarter_Summary_df = captivateiq_Attainment_by_quarter_Summary_df.merge(captivateiq_earned_Commission_Summary_df, how="inner", on=["Name", "attainment_quarter"])
captivateiq_Attainment_by_quarter_Summary_df
```

```python
#Final commission summary
captivateiq_Attainment_by_quarter_Summary_df = captivateiq_Attainment_by_quarter_Summary_df.merge(captivateiq_paid_Commission_Summary_df, how="inner", left_on=["Name", "attainment_quarter"], right_on=["Name", "payment_quarter"])
captivateiq_Attainment_by_quarter_Summary_df
```
